```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as scp
```

```python
k = [2,3,5,10,20,48]
p = np.linspace(0,1,1000)
p_r = []
for i in k:
    p_r_k = 1 - ((1-p)**(i+1) + scp.binom(i+1,1)*p*(1-p)**i)
    p_r.append((p_r_k,i))

for i in p_r:
    plt.plot(p,i[0], label = i[1])
    
# praktisch gemessene Werte


plt.legend()

for i in k:
    print("k=", i,  "p_r=",1 - ((1-0.1)**(i+1) + scp.binom(i+1,1)*0.1*(1-0.1)**i))
```
